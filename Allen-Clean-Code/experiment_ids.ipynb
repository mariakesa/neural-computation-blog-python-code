```python
from pathlib import Path
from allensdk.core.brain_observatory_cache import BrainObservatoryCache
```

```python
output_dir = '/media/maria/DATA/AllenData'
boc = BrainObservatoryCache(manifest_file=str(Path(output_dir) / 'brain_observatory_manifest.json'))
experiment_container = boc.get_experiment_containers()
```

```python
id = 660492886
exps = boc.get_ophys_experiments(experiment_container_ids=[id])
print(exps)
```

```python
exp_ids=[dct['id'] for dct in experiment_container]
print(exp_ids)
exps_more=boc.get_ophys_experiments(experiment_container_ids=exp_ids)
print(exps_more)
```

```python
import pandas as pd

df=pd.DataFrame(exps_more)

reduced_df=df[['id', 'experiment_container_id', 'session_type']]
df
reduced_df

```

```python
# Convert the grouped DataFrame to the desired dictionary format
grouped_df = df.groupby(['experiment_container_id', 'session_type'])['id'].agg(list).reset_index()
result_dict = {}
for row in grouped_df.itertuples(index=False):
    container_id, session_type, ids = row
    if container_id not in result_dict:
        result_dict[container_id] = {}
    result_dict[container_id][session_type] = ids[0]

# Print or use the result_dict as needed
print(result_dict)
```
