```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your CSV file into a pandas DataFrame
df = pd.read_csv('first_q_test.csv')

df=df.filter(regex='var_exp_clip_\d+')

# Filter rows where variance explained is above 0.5 in at least one trial
selected_rows = df[df.max(axis=1) > 0.5]
print(selected_rows)

# Calculate mean and confidence intervals
mean_values = selected_rows.mean(axis=1)
confidence_intervals = 1.96 * selected_rows.sem(axis=1)  # Assuming a 95% confidence interval

# Plotting
plt.figure(figsize=(10, 6))
plt.errorbar(x=np.arange(len(mean_values)), y=mean_values, yerr=confidence_intervals, fmt='o', capsize=5)
plt.title('Confidence Intervals of Neurons with Variance Explained > 0.5 at least once')
plt.xlabel('Neuron index')
plt.ylabel('Mean Variance Explained Value with Confidence Intervals')
plt.show()
```
