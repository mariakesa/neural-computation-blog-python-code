```python
import pandas as pd

dat=pd.read_csv('first_q_test.csv')
```

```python
clip_columns = dat.filter(regex='var_exp_clip_\d+')
dino_columns = dat.filter(regex='var_exp_dino_\d+')
vitmae_columns = dat.filter(regex='var_exp_vitmae_\d+')
print(clip_columns)
print(dino_columns)
print(vitmae_columns)
```

```python
import numpy as np
mean_clip=np.mean(clip_columns,axis=1)
mean_dino=np.mean(dino_columns,axis=1)
mean_vitmae=np.mean(vitmae_columns,axis=1)
std_clip=np.std(clip_columns,axis=1)     
std_dino=np.std(dino_columns,axis=1)
std_vitmae=np.std(vitmae_columns,axis=1)
```
